```python
import torch
import cv2
import time
from PIL import Image
# Images
```

```python
ckpt_path = '/Users/16545/Downloads/yolov3_best.pt'
model = torch.hub.load('ultralytics/yolov3', 'custom', ckpt_path)
```

```python
def yolo(im, size=640):
    g = (size / max(im.size))  # gain
    im = im.resize((int(x * g) for x in im.size), Image.Resampling.LANCZOS)  # resize
    results = model(im)  # inference
    print(results)
    results.render()  # updates results.imgs with boxes and labels
    return Image.fromarray(results.ims[0])
```

```python
img = cv2.imread("/Users/16545/Downloads/rice_1.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
inputs = Image.fromarray(img)
```

```python
yolo(inputs)
```
